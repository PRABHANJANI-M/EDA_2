```python
df["high_ph"] = np.where(df["ph"]>=7.5,1,0)
df["high_hardness"] = np.where(df["Hardness"]>df["Hardness"].mean(),1,0)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
df['chloradity'] = np.add(df['Solids'], df['Chloramines'])
threshold = df['chloradity'].median()
print("Median-based threshold:", threshold)
df['chloradity_high'] = df['chloradity'] > threshold
plt.hist(df['chloradity'], bins=20, color='skyblue', edgecolor='black')
plt.axvline(threshold, color='red', linestyle='--', label=f'Threshold = {threshold}')
plt.xlabel('Chloradity')
plt.ylabel('Count')
plt.title('Distribution of Chloradity')
plt.legend()
plt.show()
```

```python
print("Before filling mising values")
print(df.isnull().sum())
df_filled=df.fillna(df.mean())
print("After filling missing values")
print(df_filled.isnull().sum())
```

```python
df = df.dropna(axis=0)
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['ph'])
plt.title('Distribution of pH')
plt.subplot(1, 2, 2)
plt.hist(df['Hardness'])
plt.title('Distribution of Hardness')
plt.tight_layout()
plt.show()
```

```python
y_true = df['Potability'].dropna().head(10).values
y_pred = np.random.rand(10)
mse = np.mean((y_pred - y_true)**2)
grad = (2/len(y_true))*(y_pred-y_true)
print("MSE:", mse)
print("Gradient:", grad)
```

```python
counts = df['Potability'].value_counts()
plt.figure(figsize=(5,4))
plt.bar(counts.index.astype(str), counts.values, color=['skyblue','salmon'])
plt.title("Water Potability Count")
plt.xlabel("Potability")
plt.ylabel("Count")
plt.show()
```

```python
ph = df['ph'].dropna()
total_samples = len(ph)
safe_samples = ph[(ph > 6.5) & (ph < 8.5)].count()
probability = safe_samples / total_samples
print("Total samples:", total_samples)
print("Number of safe samples:", safe_samples)
print("Probability(6.5<pH <8.5):", probability)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.scatter(df['Hardness'], df['Solids'], alpha=0.6, color='purple')
plt.title("Scatter Plot: Hardness vs Solids")
plt.xlabel("Hardness")
plt.ylabel("Solids")
plt.grid(True)
plt.show()
correlation = df['Hardness'].corr(df['Solids'])
print("Correlation between Hardness and Solids:", correlation)
```

Stastical analysis.

```python
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
columns = ['ph', 'Turbidity', 'Trihalomethanes']
for column in columns:
    mean = df[column].mean()
    median = df[column].median()
    std = df[column].std()
    variance = df[column].var()
    skewness = df[column].skew()
    mode = df[column].mode()[0]
    print(f"{column} â†’ Mean: {mean:.2f}, Median: {median:.2f}, Mode: {mode:.2f}, Variance: {variance:.2f}, Skewness: {skewness:.2f}, Standard Deviation: {std:.2f}")
# columns = ['ph', 'Turbidity', 'Trihalomethanes']
for col in columns:
    sns.histplot(df[col].dropna(), bins=20, color='skyblue', kde=True, edgecolor='black')
    plt.title(f'Distribution of {col} (Skewness = {df[col].skew():.2f})')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

```

```python
import numpy as np
hardness = df['Hardness'].to_numpy()
sulfate = df['Sulfate'].to_numpy()
addition = hardness + sulfate
subtraction = hardness - sulfate
dot_product = np.dot(hardness, sulfate)
correlation = df['Hardness'].corr(df['Sulfate'])
print("Correlation between Hardness and Sulfate:", correlation)
print("Dot Product:", dot_product)
print("Addition:", addition)
print("Subtraction:", subtraction)
```
