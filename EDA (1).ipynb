```python
df["high_ph"] = np.where(df["ph"]>=7,1,0)
df["high_hardness"] = np.where(df["Hardness"]>df["Hardness"].mean(),1,0)
df
```

```python
df['Hardness'] = df['Hardness'].fillna(df['Hardness'].median())
df['Sulfate'] = df['Sulfate'].fillna(df['Sulfate'].median())
df['Hardness_plus_Sulfate'] = df['Hardness'] + df['Sulfate']
df.to_csv("water_potability.csv", index=False)
df
# print(df[['Hardness','Sulfate','Hardness_plus_Sulfate']].head())
```

```python
df['ph'] = df['ph'].fillna(df['ph'].mean())
df['Sulfate'] = df['Sulfate'].fillna(df['Sulfate'].median())
df['Trihalomethanes'] = df['Trihalomethanes'].fillna(df['Trihalomethanes'].median())
skewed_cols = ['Solids', 'Organic_carbon', 'Trihalomethanes']
for col in skewed_cols:
    df[col] = np.log1p(df[col])
    df[col + '_log'] = np.log1p(df[col])
    df[col + '_log'].hist(bins=30, edgecolor='k')
    plt.title(f'{col} after log transform')
    plt.show()
df
```

```python
y_true = df['Potability'].dropna().head(10).values
y_pred = np.random.rand(10)
mse = np.mean((y_pred - y_true)**2)
grad = (2/len(y_true))*(y_pred-y_true)
print("MSE:", mse)
print("Gradient:", grad)
```

```python
ph = df['ph']
total_samples = len(ph)
safe_samples = ph[(ph > 6.5) & (ph < 8.5)].count()
probability = safe_samples / total_samples
print("Total samples:", total_samples)
print("Number of safe samples:", safe_samples)
print("Probability(6.5<pH <8.5):", probability)
```

```python
df = pd.read_csv("water_potability.csv")
numeric_cols = ['ph','Hardness','Solids','Chloramines','Sulfate',
                'Conductivity','Organic_carbon','Trihalomethanes','Turbidity','Potability','Hardness_plus_Sulfate']
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
corr = df[numeric_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Stastical analysis.

```python
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
columns = ['ph', 'Turbidity', 'Trihalomethanes']
for column in columns:
    mean = df[column].mean()
    median = df[column].median()
    std = df[column].std()
    variance = df[column].var()
    skewness = df[column].skew()
    mode = df[column].mode()[0]
    print(f"{column}  Mean: {mean:.2f}, Median: {median:.2f}, Mode: {mode:.2f}, Variance: {variance:.2f}, Skewness: {skewness:.2f}, Standard Deviation: {std:.2f}")
```
